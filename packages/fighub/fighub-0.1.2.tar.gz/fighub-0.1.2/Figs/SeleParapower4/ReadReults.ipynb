```python
import os
import pickle

seed = 1
data_name = "CIFAR100"
Optimizer_name_list = ['SPBM-TR'] 
batch_size = 256
epoch = 50
pkl_paths = []

# 存储结果
training_loss_last = {}

for Optimizer_name in Optimizer_name_list:
    path = f"../../SeleParaPower4_Results/seed_{seed}/ResNet18/{data_name}/{Optimizer_name}/train_2000_val_0_test_10000/Batch_size_{batch_size}/epoch_{epoch}/2025-08-30_17-14-14"
    pkl_files = [f for f in os.listdir(path) if f.endswith(".pkl")]
    
    for f in pkl_files:
        full_path = os.path.join(path, f)
        pkl_paths.append(full_path)

# 逐个读取 pkl 文件，并存储最后一个 training_loss
for pkl_file in pkl_paths:
    with open(pkl_file, 'rb') as f:
        data = pickle.load(f)
        # 用文件名作为字典键，保存最后一个训练损失
        training_loss_last[os.path.basename(pkl_file)] = data["training_loss"][-1]

results = {
    'training_loss': [],
    'data_name':  "CIFAR100"
}
for i in [4**i for i in range(-4, 5)]:
    results['training_loss'].append(training_loss_last[f"SPBM-TR_M_1e-05_delta_{i}_cutting_number_10.pkl"])


with open('SPBM-TR_M_1e-05_delta_0_cutting_number_10.pkl', 'wb') as f:
    pickle.dump(results, f)

```
